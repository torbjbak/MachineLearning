```python
import pandas as pd

data = pd.read_csv("agaricus-lepiota.data", 
                    names=['edibility','cap-shape','cap-surface','cap-color',
                        'bruises?','odor','gill-attachment','gill-spacing',
                        'gill-size','gill-color','stalk-shape','stalk-root',
                        'stalk-surface-above-ring','stalk-surface-below-ring',
                        'stalk-color-above-ring','stalk-color-below-ring','veil-type',
                        'veil-color','ring-number','ring-type','spore-print-color','population','habitat']
)

data = data[1:]
```

```python
import matplotlib.pyplot as plt

edible = data[data['edibility'] == 'e']
poisonous = data[data['edibility'] == 'p']

e_hab_count = edible['habitat'].value_counts().to_frame().rename(columns={"habitat": "Edible"})
p_hab_count = poisonous['habitat'].value_counts().to_frame().rename(columns={"habitat": "Poisonous"})

plot_data = pd.DataFrame(pd.concat([e_hab_count, p_hab_count], axis=1))

plot_data.plot(kind="bar", stacked=True)
plt.xticks(rotation=0, horizontalalignment="center")
plt.xlabel("Habitat type")
plt.ylabel("Mushroom observations")
```

As illustrated by the stacked bar chart above, you can see that there is a certain correlation between habitat type and edibility. For example, habitat 'm' and 'w' are almiost entirely representet by edible mushrooms, while habitat 'p'  is heavily overrepresented by poisonous mushrooms.

```python
dummies = pd.get_dummies(data)

plt.figure()
fig = plt.gcf()
fig.set_size_inches(12, 8)
plt.spy(dummies, aspect='auto', markersize=0.4)
plt.show()
```
